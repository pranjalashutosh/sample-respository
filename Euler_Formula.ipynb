```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.arange(0, 2*np.pi,0.01)
y=np.sin(4*x)# 1st wave
y1=np.sin(x) # 2nd wave with different frequency
plt.plot(x,y) 
plt.plot(x,y1)
plt.xlabel('x - axis')
plt.ylabel('y - axis')
plt.show()
```

```python
# Plotting the added version of the two sine wave with different frequencies presented above:
gt=y+y1
plt.plot(x,gt) 
plt.show()
# This is to show that we can add Sine waves together to get a new complex wave. 
```

```python
# Defining a function for 'e'. 
import math
def exp(x):
    return sum([
        x**n/math.factorial(n)
        for n in range(0,100)
    ])
exp(1)
```

```python
cosine=[]
sine=[]
a1=np.arange(0,7,0.1)
for i in a1:
    cosine.append(np.cos(i))
    sine.append(np.sin(i))

plt.plot(cosine,sine) # this will plot cos(o) + isin(o), this also shows the rotational property. 
plt.show()
```
